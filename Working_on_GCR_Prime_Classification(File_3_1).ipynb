```python
import pandas as pd

# Load datasets
df_routes = new_df
df_flights = pd.read_csv("lets.csv")

# Clean column names
df_routes.columns = df_routes.columns.str.strip().str.lower().str.replace(" ", "_")
df_flights.columns = df_flights.columns.str.strip().str.lower().str.replace(" ", "_")

# Filter for available flights only
available_flights = df_flights[df_flights["flight_available_(updated)"].str.lower() == "yes"]

# Build lookup dictionary: (origin_city, destination_city) → row
flight_lookup = {}
for _, row in available_flights.iterrows():
    o = row['origin_city'].strip().lower()
    d = row['destination_city'].strip().lower()
    flight_lookup[(o, d)] = row

# Function to generate flow path with correct logic
def get_flow_path(origin, destination, lookup):
    key = (origin.strip().lower(), destination.strip().lower())
    if key not in lookup:
        return None

    row = lookup[key]

    path = [
        row['origin_city'],
        row['origin_hub_city'],
        row['origin_airport_city'],
        row['destination_airport_city'],
        row['destination_hub_city'],
        row['destination_city']
    ]

    # Remove consecutive duplicates
    deduped_path = [path[0]]
    for city in path[1:]:
        if city != deduped_path[-1]:
            deduped_path.append(city)

    # Add travel mode symbols: ✈️ for flight segment only
    full_path_with_icons = []
    for i in range(len(deduped_path) - 1):
        current_city = deduped_path[i]
        next_city = deduped_path[i + 1]

        if current_city == row['origin_airport_city'] and next_city == row['destination_airport_city']:
            full_path_with_icons.append(f"{current_city} ✈️")
        else:
            full_path_with_icons.append(f"{current_city} 🛣️")

    full_path_with_icons.append(deduped_path[-1])  # Add the last city

    return " ".join(full_path_with_icons)

# Function to build route DataFrame by category
def build_route_df(category):
    filtered = df_routes[df_routes["aeroplane_category"].str.upper() == category.upper()]
    records = []

    for _, row in filtered.iterrows():
        origin = row['origin_city']
        destination = row['destination_city']
        flow = get_flow_path(origin, destination, flight_lookup)

        if flow:
            records.append({
                'origin_city': origin,
                'destination_city': destination,
                'flow_path': flow
            })

    return pd.DataFrame(records)

# Build GCR and PRIME route dataframes
gcr_df = build_route_df("GCR")
prime_df = build_route_df("PRIME")

# Save to CSVs
gcr_df.to_csv("GCR_Flight_Flows.csv", index=False)
prime_df.to_csv("PRIME_Flight_Flows.csv", index=False)

# Print preview
print("✅ GCR Preview:")
print(gcr_df.head())

print("\n✅ PRIME Preview:")
print(prime_df.head())
```

```python
import pandas as pd

# Load input CSVs
shipment_df = pd.read_csv("lets.csv")
flight_routes_df = new_df

# Standardize column names
shipment_df.columns = shipment_df.columns.str.strip().str.lower().str.replace(" ", "_")
flight_routes_df.columns = flight_routes_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Map of (origin_airport, dest_airport) → [GCR, PRIME]
flight_type_map = {}
for _, row in flight_routes_df.iterrows():
    orig = row['origin_city'].strip().lower()
    dest = row['destination_city'].strip().lower()
    category = row['aeroplane_category'].strip().upper()
    flight_type_map.setdefault((orig, dest), set()).add(category)

# Containers for GCR and PRIME outputs
gcr_data = []
prime_data = []

# Process each shipment route
for _, row in shipment_df.iterrows():
    if str(row['flight_available_(updated)']).strip().lower() != 'yes':
        continue  # Only consider available flights

    # Clean city names
    origin_city = row['origin_city'].strip()
    origin_hub = row['origin_hub_city'].strip()
    origin_airport = row['origin_airport_city'].strip()
    dest_airport = row['destination_airport_city'].strip()
    dest_hub = row['destination_hub_city'].strip()
    destination_city = row['destination_city'].strip()

    # Build full path
    raw_path = [origin_city, origin_hub, origin_airport, dest_airport, dest_hub, destination_city]

    # Remove consecutive duplicates
    cleaned_path = [raw_path[0]]
    for city in raw_path[1:]:
        if city != cleaned_path[-1]:
            cleaned_path.append(city)

    # Add icons (✈️ between airport cities, 🛣️ otherwise)
    final_path = []
    for i in range(len(cleaned_path) - 1):
        city_from = cleaned_path[i]
        city_to = cleaned_path[i + 1]
        if city_from == origin_airport and city_to == dest_airport:
            final_path.append(f"{city_from} ✈️")
        else:
            final_path.append(f"{city_from} 🛣️")
    final_path.append(cleaned_path[-1])  # Add final city

    flow_path_str = " ".join(final_path)

    # Determine flight type
    key = (origin_airport.strip().lower(), dest_airport.strip().lower())
    categories = flight_type_map.get(key, set())

    if "GCR" in categories:
        gcr_data.append({
            "origin_city": origin_city,
            "destination_city": destination_city,
            "flow_path": flow_path_str
        })
    if "PRIME" in categories:
        prime_data.append({
            "origin_city": origin_city,
            "destination_city": destination_city,
            "flow_path": flow_path_str
        })

# Create DataFrames and save
pd.DataFrame(gcr_data).to_csv("GCR_Flight_Flows_Final.csv", index=False)
pd.DataFrame(prime_data).to_csv("PRIME_Flight_Flows_Final.csv", index=False)

print("✅ GCR and PRIME flight flows saved successfully.")
```

```python
import pandas as pd

# STEP 1: Load shipment dataset
df = pd.read_csv("shipment_data.csv")

# STEP 2: Clean and prepare data
df["Weights"] = df["Weights"].fillna(0)
df["Conno"] = df["Conno"].fillna(0)

# STEP 3: Create a Flight_Mode column
def get_mode(row):
    if row["Flight_Mode"] == "NO_FLIGHT":
        return "NO_FLIGHT"
    elif row["Flight_Mode"] == "PRIME+GCR":
        return "PRIME+GCR"
    elif row["Flight_Mode"] == "PRIME":
        return "PRIME"
    elif row["Flight_Mode"] == "GCR":
        return "GCR"
    else:
        return "UNKNOWN"

df["Flight_Mode"] = df.apply(get_mode, axis=1)

# STEP 4: Save records with NO_FLIGHT mode (ODs that need alternate routing)
no_flight_df = df[df["Flight_Mode"] == "NO_FLIGHT"]
no_flight_df.to_csv("no_flight_OD_pairs.csv", index=False)
print("✅ Saved OD pairs with NO_FLIGHT to 'no_flight_OD_pairs.csv'")

# STEP 5: Aggregate volume at airport level
airport_volume = (
    df.groupby(["Origin Airport City", "Destination Airport City"])
    .agg(
        Total_Weight=("Weights", "sum"),
        Total_ChargedWeight=("Conno", "sum"),
        Total_Shipments=("Conno", "sum"),
        Unique_Products=("New Product GCP", pd.Series.nunique),
        Num_OD_Branches=("Destination Branch", pd.Series.nunique),
        Modes=("Flight_Mode", lambda x: ", ".join(sorted(set(x))))
    )
    .reset_index()
)

# STEP 6: Save airport-to-airport volume matrix
airport_volume.to_csv("airport_to_airport_volume.csv", index=False)
print("✅ Aggregated airport volumes saved to 'airport_to_airport_volume.csv'")
```
