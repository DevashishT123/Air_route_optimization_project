```python
import pandas as pd

# Load datasets
shipment_df = df1
mapping_df = pd.read_csv("/content/branch-hub-airport-mapping.csv")

# Clean whitespace
shipment_df['Origin Branch'] = shipment_df['Origin Branch'].str.strip()
shipment_df['Destination Branch'] = shipment_df['Destination Branch'].str.strip()
mapping_df['Branch'] = mapping_df['Branch'].str.strip()
mapping_df['Hub'] = mapping_df['Hub'].str.strip()

# === Create branch-based lookup ===
branch_lookup = mapping_df.set_index('Branch')[['Hub', 'Hub City', 'Final_Airport_City']].to_dict(orient='index')

# === Create hub-name-only lookup for fallback ===
mapping_df['Hub Name Only'] = mapping_df['Hub'].str.split('-', n=1).str[1]

# Drop duplicates to avoid index conflict
hub_name_only_df = mapping_df[['Hub Name Only', 'Hub', 'Hub City', 'Final_Airport_City']].drop_duplicates(subset='Hub Name Only')

# Set index for lookup
hub_lookup = hub_name_only_df.set_index('Hub Name Only')[['Hub', 'Hub City', 'Final_Airport_City']].to_dict(orient='index')

# === Mapping logic ===
def get_mapping(branch_name):
    branch_name = branch_name.strip()

    # First try direct branch match
    if branch_name in branch_lookup:
        return branch_lookup[branch_name]

    # If not found, try to match as hub name only (after removing code)
    name_only = branch_name.split('-', 1)[-1]
    if name_only in hub_lookup:
        return hub_lookup[name_only]

    return {'Hub': None, 'Hub City': None, 'Final_Airport_City': None}

# Apply mapping to Origin and Destination
origin_info = shipment_df['Origin Branch'].apply(get_mapping).apply(pd.Series)
destination_info = shipment_df['Destination Branch'].apply(get_mapping).apply(pd.Series)

# Rename columns accordingly
origin_info.columns = ['Origin Hub', 'Origin Hub City', 'Origin Airport City']
destination_info.columns = ['Destination Hub', 'Destination Hub City', 'Destination Airport City']

# Combine everything
final_df = pd.concat([shipment_df, origin_info, destination_info], axis=1)

# Save
final_df.to_csv("final_mapped_shipment.csv", index=False)
print(" Mapping complete. Saved as 'final_mapped_shipment.csv'")
```

```python
# Load flight map
flight_map = pd.read_csv("/content/flight_connection.csv")

flight_map['origin_city'] = flight_map['origin_city'].str.upper().str.strip()
flight_map['destination_city'] = flight_map['destination_city'].str.upper().str.strip()

# Create sets
prime_set = set(zip(flight_map[flight_map['Aeroplane category'] == 'PRIME']['origin_city'],
                    flight_map[flight_map['Aeroplane category'] == 'PRIME']['destination_city']))
gcr_set = set(zip(flight_map[flight_map['Aeroplane category'] == 'GCR']['origin_city'],
                  flight_map[flight_map['Aeroplane category'] == 'GCR']['destination_city']))

# Function to get mode
def get_flight_mode(o_airport, d_airport):
    pair = (str(o_airport).upper().strip(), str(d_airport).upper().strip())
    if pair in prime_set and pair in gcr_set:
        return "PRIME+GCR"
    elif pair in prime_set:
        return "PRIME"
    elif pair in gcr_set:
        return "GCR"
    else:
        return "NO_FLIGHT"

# Apply
final_df['Flight_Mode'] = final_df.apply(lambda row: get_flight_mode(row['Origin Airport City'], row['Destination Airport City']), axis=1)
```

```python
import pandas as pd
from geopy.distance import geodesic
from itertools import product

# Load city coordinates
city_df = pd.read_csv('/content/city_coordinate.csv')  # <-- Upload this file before running

# Clean and prepare city coordinates
city_df = city_df.dropna(subset=['City', 'Latitude', 'Longitude'])
city_df['City'] = city_df['City'].str.strip().str.upper()

# Create all possible OD pairs (excluding same-city pairs)
city_pairs = pd.DataFrame(list(product(city_df['City'], repeat=2)), columns=['Origin', 'Destination'])
city_pairs = city_pairs[city_pairs['Origin'] != city_pairs['Destination']]  # Remove same city

# Merge coordinates for both cities
city_pairs = city_pairs.merge(city_df.rename(columns={'City': 'Origin', 'Latitude': 'Origin_Lat', 'Longitude': 'Origin_Lon'}), on='Origin')
city_pairs = city_pairs.merge(city_df.rename(columns={'City': 'Destination', 'Latitude': 'Dest_Lat', 'Longitude': 'Dest_Lon'}), on='Destination')

# Calculate geodesic distance (in km) and estimated flight time (in minutes)
def estimate_flight_time(row):
    origin_coords = (row['Origin_Lat'], row['Origin_Lon'])
    dest_coords = (row['Dest_Lat'], row['Dest_Lon'])
    distance_km = geodesic(origin_coords, dest_coords).kilometers
    flight_speed_kmph = 750  # Fast commercial aircraft
    time_hours = distance_km / flight_speed_kmph
    return round(time_hours * 60, 1)  # in minutes

city_pairs['Estimated_Flight_Time_Minutes'] = city_pairs.apply(estimate_flight_time, axis=1)

# Optional: filter by max flight time (e.g., under 3 hours)
# city_pairs = city_pairs[city_pairs['Estimated_Flight_Time_Minutes'] <= 180]

# Output: city_pairs now contains all estimated flight times
# You can save this to a file or join with your volume dataset
city_pairs[['Origin', 'Destination', 'Estimated_Flight_Time_Minutes']].to_csv('flight_time_matrix.csv', index=False)
print(" Flight time matrix saved as 'flight_time_matrix.csv'")
```

```python
import pandas as pd
import numpy as np

# Load datasets
flight_df = pd.read_csv("/content/flight_time_matrix.csv")
road_df = pd.read_csv("/content/city_distances_google.csv")
volume_df = pd.read_csv("/content/new_to_look.csv")

# Filter NO_FLIGHT rows where origin  destination
volume_nf = volume_df[
    (volume_df['Flight_Mode'] == 'NO_FLIGHT') &
    (volume_df['Origin Airport City'] != volume_df['Destination Airport City'])
].copy()

# Clean and prepare road duration
road_df = road_df.dropna(subset=['duration_hr']).copy()
road_df['Road_Duration_Minutes'] = road_df['duration_hr'] * 60

# Rename columns for merge clarity
flight_df.rename(columns={'Origin': 'Origin_Airport', 'Destination': 'Midpoint_Airport',
                          'Estimated_Flight_Time_Minutes': 'Flight_Time_Minutes'}, inplace=True)
road_df.rename(columns={'origin': 'Midpoint_Airport', 'destination': 'Destination_Airport'}, inplace=True)

# Step 1: Cross join volume with possible midpoints from flight_df
# Get all possible midpoints for each origin
midpoints_df = flight_df[['Origin_Airport', 'Midpoint_Airport', 'Flight_Time_Minutes']]

# Merge with volume_nf on Origin_Airport
merged = volume_nf.merge(midpoints_df, left_on='Origin Airport City', right_on='Origin_Airport', how='left')

# Merge with road_df to get road time from midpoint to destination
merged = merged.merge(
    road_df[['Midpoint_Airport', 'Destination_Airport', 'Road_Duration_Minutes']],
    left_on=['Midpoint_Airport', 'Destination Airport City'],
    right_on=['Midpoint_Airport', 'Destination_Airport'],
    how='inner'
)

# Compute total time
merged['Total_Time_Minutes'] = merged['Flight_Time_Minutes'] + merged['Road_Duration_Minutes']

# Rename columns and keep relevant
merged.rename(columns={
    'Origin Airport City': 'Origin',
    'Destination Airport City': 'Destination',
    'Weights': 'Weight',
    'Conno': 'Conno'
}, inplace=True)

# Normalize for scoring
merged['Normalized_Weight'] = merged['Weight'] / merged['Weight'].max()
merged['Normalized_Time'] = merged['Total_Time_Minutes'] / merged['Total_Time_Minutes'].max()

# Scoring formula
alpha = 0.6
beta = 0.4
merged['Path_Score'] = (
    alpha * merged['Normalized_Weight'] +
    beta * (1 - merged['Normalized_Time'])
)

# Select best path per OD
merged['OD_Key'] = merged['Origin'] + "->" + merged['Destination']
best_paths_df = merged.sort_values(by=['OD_Key', 'Path_Score'], ascending=[True, False]) \
                      .drop_duplicates('OD_Key', keep='first')

# Save results
best_paths_df.to_csv("ranked_indirect_paths.csv", index=False)

print(" Optimized indirect paths computed. Sample:")
print(best_paths_df[['Origin', 'Midpoint_Airport', 'Destination', 'Total_Time_Minutes', 'Weight', 'Path_Score']].head(10))
```

```python
import pandas as pd
import networkx as nx
from tqdm import tqdm

# Load datasets
flight_df = pd.read_csv("/content/flight_time_matrix.csv")
road_df = pd.read_csv("/content/city_distances_google.csv")
volume_df = pd.read_csv("/content/new_to_look.csv")

# Preprocess data
road_df = road_df.dropna(subset=['duration_hr']).copy()
road_df['Road_Duration_Minutes'] = road_df['duration_hr'] * 60
flight_df['Flight_Time_Minutes'] = flight_df['Estimated_Flight_Time_Minutes']

# Create graph
G = nx.DiGraph()

# Add flight edges
for _, row in flight_df.iterrows():
    G.add_edge(row['Origin'], row['Destination'],
               type='flight',
               time=row['Flight_Time_Minutes'])

# Add road edges
for _, row in road_df.iterrows():
    G.add_edge(row['origin'], row['destination'],
               type='road',
               time=row['Road_Duration_Minutes'])

# Filter volume_df to only OD pairs with no direct flight
volume_nf = volume_df[
    (volume_df['Flight_Mode'] == 'NO_FLIGHT') &
    (volume_df['Origin Airport City'] != volume_df['Destination Airport City'])
].copy()

# Index lookups
flight_lookup = flight_df.set_index(['Origin', 'Destination'])['Flight_Time_Minutes'].to_dict()
road_lookup = road_df.set_index(['origin', 'destination'])['Road_Duration_Minutes'].to_dict()

results = []

# Efficient search for 1-hop indirect paths
for _, row in tqdm(volume_nf.iterrows(), total=volume_nf.shape[0]):
    origin_air = row['Origin Airport City']
    dest_air = row['Destination Airport City']
    weight = row['Weights']
    conno = row['Conno']

    # Midpoint airports that are directly reachable by air from origin
    possible_mids = flight_df[flight_df['Origin'] == origin_air]['Destination'].unique()

    for mid in possible_mids:
        flight_leg = (origin_air, mid)
        road_leg = (mid, dest_air)

        if flight_leg in flight_lookup and road_leg in road_lookup:
            f_time = flight_lookup[flight_leg]
            r_time = road_lookup[road_leg]
            total_time = f_time + r_time

            results.append({
                'Origin': origin_air,
                'Midpoint_Airport': mid,
                'Destination': dest_air,
                'Flight_Time_Minutes': f_time,
                'Road_Time_Minutes': r_time,
                'Total_Time_Minutes': total_time,
                'Weight': weight,
                'Conno': conno
            })

# Convert to DataFrame
paths_df = pd.DataFrame(results)

# Normalize
paths_df['Normalized_Weight'] = paths_df['Weight'] / paths_df['Weight'].max()
paths_df['Normalized_Time'] = paths_df['Total_Time_Minutes'] / paths_df['Total_Time_Minutes'].max()

# Score with tunable weights
alpha = 0.6  # preference toward volume
beta = 0.4   # preference toward shorter time

paths_df['Path_Score'] = (
    alpha * paths_df['Normalized_Weight'] +
    beta * (1 - paths_df['Normalized_Time'])
)

# Rank by OD pair
paths_df['OD_Key'] = paths_df['Origin'] + "->" + paths_df['Destination']
paths_df = paths_df.sort_values(by=['OD_Key', 'Path_Score'], ascending=[True, False])

# Keep best path per OD
best_paths = paths_df.drop_duplicates(subset='OD_Key', keep='first')

# Save output
best_paths.to_csv("best_indirect_air_paths.csv", index=False)
print(" Saved best alternate air-road paths to 'best_indirect_air_paths.csv'")
```
